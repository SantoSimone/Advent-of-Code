```
# PART ONE
@lru_cache(1001)
def hash_gen(code):
  code = hashlib.md5(code.encode()).hexdigest()
  return code

def is_key(salt, i):
  # First rule: same char repeating 3 times in MD5 hash
  match = re.search(r'(.)\1\1', hash_gen(f'{salt}{i}'))
  if match:
    ch = match.group(1) * 5
    # Second rule: one of the next 1000 MD5 hashes must contain a char repeating 5 times
    if any([True if ch in hash_gen(f'{salt}{off}') else False for off in range(i + 1 , i + 1001)]):
      return 1 
    else:
      return 0
  
  return 0

def part_one(salt):
  i = 0
  keys = 0
  while keys < 64:
    keys += is_key(salt, i)
    i += 1
  
  return i - 1

salt = 'cuanljph'

print(f'64-th key is generated by index: {part_one(salt)}')
```
