```
import urllib.request
import numpy as np

# It's day 8
day = 8

# Load file from url
txt = urllib.request.urlopen(f'https://raw.githubusercontent.com/SantoSimone/Advent-of-Code/master/2016/input_files/input{day}.txt').read().decode('utf-8')
# Split file into lines
lines = txt.split('\n')[:-1]
lines
```

```
# PART ONE 
screen = np.zeros((6, 50))

def exec(cmd, params, screen):
  if cmd == 'rect':
    w, h = params[0].split('x')
    screen[:int(h), :int(w)] = 1

  elif cmd == 'rotate' and params[0] == 'column':
    col = int(params[1][2:])
    shift = int(params[-1])
    tmp = copy.deepcopy(screen[:, col])
    screen[:shift, col] = tmp[(6 - shift):]
    screen[shift:, col] = tmp[:(6 - shift)]

  elif cmd == 'rotate' and params[0] == 'row':
    row = int(params[1][2:])
    shift = int(params[-1])
    tmp = copy.deepcopy(screen[row, :])
    screen[row, :shift] = tmp[(50 - shift):]
    screen[row, shift:] = tmp[:(50 - shift)]

  return screen

for line in lines:
  chars = line.split(' ')
  screen = exec(chars[0], chars[1:], screen)

print(f'Final number of lights: {np.sum(screen, (0,1))}')

# PART TWO
for i in range(screen.shape[0]):
  for j in range(screen.shape[1]):
    if screen[i, j]:
      print('#', end='')
    else:
      print('-', end='')
  print('-')
```
